```python
import pandas as pd
```

## Загрузка данных с предыдущего задания

```python
data = pd.read_json("../ex02/auto.json")

data.set_index("CarNumber", inplace=True)

data
```

## Выборки

- `Fines` больше 2100

```python
data_fines_high_2100 = data[data["Fines"] > 2100]

data_fines_high_2100
```

- `Fines` больше 2100 и `Refund` равен 2

```python
data_fines_high_2100_2 = data[(data["Fines"] > 2100) & (data["Refund"] == 2)]

data_fines_high_2100_2
```

- `Model` входит в список: `['Focus', 'Corolla']`

```python
data_model = data[data["Model"].isin(["Focus", "Corolla"])]

data_model 
```

- `CarNumber` входит в список:
`['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']`

```python
data_car_number = data[data.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]

data_car_number
```

## Делаем агрегации по марке и модели

- Медиана значений `Fines` сгруппированных по `Make`

```python
data_median_fines_by_make = data.groupby('Make')['Fines'].median()

data_median_fines_by_make
```

- Медиана `Fines` по `Make` и `Model`

```python
data_median_fines_by_make_model = data.groupby(['Make', 'Model'])['Fines'].median()

data_median_fines_by_make_model
```

- Количество значений `Fines` по `Make` и `Model`

```python
data_count_fines_by_make_model = data.groupby(['Make', 'Model'])['Fines'].count()

data_count_fines_by_make_model
```

- *Min* и *Max* `Fines` по `Make` и `Model`

- `agg()` -> агрегация применяет функции `min` и `max` к значениям `Fines` в каждой группе

```python
data_min_max_fines_by_make_model = data.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])

data_min_max_fines_by_make_model
```

- Стандартное отклонение `Fines` по `Make` и `Model`

- `['Fines'].std()` -> для каждой группы вычисляет стандартное отклонение значений в столбце `'Fines'`. Стандартное отклонение показывает, насколько значения штрафов в группе отклоняются от их среднего значения. Игнорирует `NaN`

```python
data_std_fines_by_make_model = data.groupby(['Make', 'Model'])['Fines'].std()

data_std_fines_by_make_model
```

## Агрегация по номеру машины

- Отсортируем номера машин по количеству штрафов `Fines` в порядке убывания чтобы найти самых частых нарушителей

```python
data_fines_grouped_by_car_number = data.groupby('CarNumber')['Fines'].count().sort_values(ascending=False)

data_fines_grouped_by_car_number
```

- Все строки относящиеся к машине с максимальным количеством штрафов

```python
top_car_number_1 = data_fines_grouped_by_car_number.idxmax()

data_top_car_number_1 = data[data.index == top_car_number_1]
```

```python
top_car_number_1
```

```python
data_top_car_number_1
```

- Сортируем номера машин по сумме штрафов в порядке убывания тобы найти кто заплатил больше всего

```python
data_fines_sum_by_car_number = data.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)

data_fines_sum_by_car_number
```

- Те кто заплатили больше всего штрафов

```python
top_car_number_2 = data_fines_sum_by_car_number.idxmax()

data_top_car_number_2 = data[data.index == top_car_number_2]
```

```python
top_car_number_2
```

```python
data_top_car_number_2
```

## Есть ли номера машин, связанные с разными моделями?

- `nunique()` -> Подсчитывает количество уникальных значений в столбце `Model` для каждой группы

```python
data_car_number_models = data.groupby('CarNumber')['Model'].nunique()

data_car_number_models[data_car_number_models > 1]

data_car_number_models
```
