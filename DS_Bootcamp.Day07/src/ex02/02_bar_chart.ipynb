```python
import pandas as pd
import sqlite3
```

```python
data = sqlite3.connect("../data/checking-logs.sqlite")
```

## Берем данные с `checker`

- Преобразуем `timestamp` в `datetime64[ns]`

```python
sql_request1 = """
SELECT timestamp FROM checker
WHERE uid LIKE 'user_%'
"""

dataframe_checker = pd.read_sql(sql_request1, data, parse_dates="timestamp")
```

## Создаем новый столбец `day_time`

```python
dataframe_checker["day_time"] = pd.cut(dataframe_checker["timestamp"].dt.hour, bins = [-1, 4, 10, 17, 24], labels = ["night", "morning", "afternoon", "evening"])
```

```python
dataframe_checker
```

## Делим на категории по `day_time`

```python
daily_commits = dataframe_checker.groupby(dataframe_checker['timestamp'].dt.date)['day_time'].value_counts().unstack(1)
```

```python
daily_commits
```

```python
daily_commits.plot(kind='bar', figsize=(15, 8), fontsize=8, stacked=True, colormap="viridis", alpha=0.7, title="Commits per day")
```

## Когда наши пользователи обычно делают коммиты лабораторных — ночью, утром, днём или вечером?

```python
total_commits = dataframe_checker["day_time"].value_counts()
```

```python
total_commits.nlargest(2)
```

## Какой день имеет

    - наибольшее количество коммитов
    
    - и при этом количество коммитов вечером больше, чем днём

```python
daily_commits[daily_commits["evening"] > daily_commits["afternoon"]].sum(axis=1).idxmax()
```
