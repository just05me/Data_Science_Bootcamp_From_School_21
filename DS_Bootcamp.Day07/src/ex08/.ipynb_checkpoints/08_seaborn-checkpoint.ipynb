```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
data = sqlite3.connect("../data/checking-logs.sqlite")
```

```python
sql_request1 = """
SELECT uid, timestamp, numTrials FROM checker
WHERE uid LIKE "user_%" AND status = "ready" AND
labname = "project1"
"""

dataframe_checker = pd.read_sql(sql_request1, data, parse_dates='timestamp')
```

```python
dataframe_checker
```

```python
sns.set_theme()

sns.relplot(data=dataframe_checker, x='timestamp', y='numTrials', hue='uid', linewidth=3, legend='full', kind='line', height=10, aspect=1.5)

plt.title('Dynamic of commits per user in project1')
```
